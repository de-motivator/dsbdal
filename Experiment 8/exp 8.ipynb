```python
def fun1(value):
    if (value == "male"):
        return 1
    else:
        return 0
```

```python
def fun2(value):
    if (value == 'S'):
        return 0
    elif (value == 'C'):
        return 1
    elif (value == 'Q'):
        return 2
    else:
        return 0
```

```python
df["Sex"] = df["Sex"].apply(fun1)
```

```python
df["Embarked"] = df["Embarked"].apply(fun2)
```

```python
df.isna().sum()
```

```python
df = df.drop("Cabin", axis=1)
```

```python
df.shape
```

```python
non_numeric_cols = df.select_dtypes(include=[object]).columns.tolist()
df_corr = df.drop(columns=non_numeric_cols)
corr_matrix = df_corr.corr(method='spearman')
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# 'Survived' is the label
```

```python
sns.countplot(df["Survived"])
plt.show()
```

```python
df_long = df.melt(id_vars="Pclass", value_name="count", var_name="Survived")
sns.countplot(x="Pclass", hue="Survived", data=df_long)  
plt.show()
```

```python
sex_survival_counts = df.groupby("Sex")["Survived"].value_counts().unstack(fill_value=0)
sex_survival_counts.plot(kind="bar", stacked=True, colormap="Set2")
plt.xlabel("Sex")
plt.ylabel("Passenger Count")
plt.title("Passenger Survival Count by Sex (from CSV)")
plt.legend(title="Survived")
plt.show()
```

```python
embarked_survival_counts = df.groupby("Embarked")["Survived"].value_counts().unstack(fill_value=0)
embarked_survival_counts.plot(kind="bar", colormap="Set3")
plt.xlabel("Embarked Port")
plt.ylabel("Passenger Count")
plt.title("Passenger Survival Count by Embarked Port (from CSV)")
plt.legend(title="Survived")
plt.show()
```
